```python
# grade_report_list_dd = None

# this_ts = datetime.strptime(os.path.basename('/Users/igahyeon/IoT-Acceleration-Sensor-Data-Analysis/dataset/sensor_grade_2305021700_2305021800_UTC.csv').split('_')[2], 
#                             "%y%m%d%H%M"
#                             ).replace(tzinfo=timezone.utc)\
#                             .astimezone(tz=timezone(timedelta(hours=9)))
# df = dd.read_csv('/Users/igahyeon/IoT-Acceleration-Sensor-Data-Analysis/dataset/sensor_grade_2305021700_2305021800_UTC.csv', dtype={'latitude': 'float32', 'longitude': 'float32'}
#                 ).assign(ts=lambda x: this_ts)
# df['usim'] = df['usim'].map(lambda x: f"{x:011d}")

# if grade_report_list_dd is None:
#     grade_report_list_dd = df
# else:
#     grade_report_list_dd = dd.concat([grade_report_list_dd, df])
    
# grade_report_list_dd

#----------------------------------
report_list = ['/Users/igahyeon/IoT-Acceleration-Sensor-Data-Analysis/dataset/sensor_grade_2305021700_2305021800_UTC.csv'
              , '/Users/igahyeon/IoT-Acceleration-Sensor-Data-Analysis/dataset/sensor_grade_2305022300_2305030000_UTC.csv'
              , '/Users/igahyeon/IoT-Acceleration-Sensor-Data-Analysis/dataset/sensor_grade_2305030500_2305030600_UTC.csv'
              , '/Users/igahyeon/IoT-Acceleration-Sensor-Data-Analysis/dataset/sensor_grade_2305030900_2305031000_UTC.csv']

grade_report_list_dd = None

for r in tqdm(report_list):
    this_ts = datetime.strptime(os.path.basename(r).split('_')[2], 
                                "%y%m%d%H%M"
                               ).replace(tzinfo=timezone.utc)\
                                .astimezone(tz=timezone(timedelta(hours=9)))
    df = dd.read_csv(r, dtype={'latitude': 'float32', 'longitude': 'float32'}
                    ).assign(ts=lambda x: this_ts)
    df['usim'] = df['usim'].map(lambda x: f"{x:011d}")
    
    if grade_report_list_dd is None:
        grade_report_list_dd = df
    else:
        grade_report_list_dd = dd.concat([grade_report_list_dd, df])
    
grade_report_list_dd
```

```python
grade_report_list_dd['ts_time'] = grade_report_list_dd['ts'].map(lambda x: x.time())
grade_report_list_dd['ts_weekday'] = grade_report_list_dd['ts'].map(lambda x: x.weekday())
grade_report_list_df = grade_report_list_dd.astype({'ts_weekday': 'int32'}).compute()
```

```python
sensor_grade_grouped = grade_report_list_df.groupby(by=['usim', 'ts_weekday','ts_time'])
sensor_grade_grouped.size()

temp_cnt = {'usim': [], 'ts_time': [], 'ts_weekday': [], 'health': []}

l = len(sensor_grade_grouped)
i = 1
for key, data in sensor_grade_grouped:    
    print(f"{i} of {l}", end='\r')
    i += 1
    
    data_quality_count = data.groupby(by=['health']).count()
    
    temp_cnt['usim'].append(key[0])
    temp_cnt['ts_time'].append(key[1])
    temp_cnt['ts_weekday'].append(key[2])
    
    if 'red' in data_quality_count.index:
        # red가 하나라도 있으면 해당 센서는 red 등급
        temp_cnt['health'].append('red')
    else:
        # 아니면 green이랑 yellow 중에 많은거로 할당
        idxmax = data_quality_count.idxmax()[0]
        temp_cnt['health'].append(idxmax)
    
grade_report_list_df = pd.DataFrame(temp_cnt)
print(grade_report_list_df)
grade_report_list_df.describe()
```

```python
## 중복센서 찾기

# 전체 샘플 수 구하기
samples_all_day = len(report_list) # 각 csv 파일에는 3개 축이 각각 있으므로 usim이 3개씩 있기 때문
print(len(report_list), samples_all_day)

group_sensors_all_day = grade_report_list_df.groupby(by=['usim']).size().to_dict()


duplicated_sensors_list = []
for usim, cnt in group_sensors_all_day.items():
    if cnt == samples_all_day: # 개수 같은 usim만 선택
        duplicated_sensors_list.append(usim)
        
duplicated_sensors_df = pd.DataFrame(duplicated_sensors_list, columns=['usim'])
duplicated_sensors_df
```

```python
duplicated_sensors_df = duplicated_sensors_df.astype({ 'usim' : int })
sensor_list = sensor_list.astype({ 'usim' : int })
grade_report_list_df = grade_report_list_df.astype({ 'usim' : int })
duplicated_grade_report_df = pd.merge(sensor_list, duplicated_sensors_df,
                                 how='right', on=['usim'])

duplicated_grade_report_df = pd.merge(grade_report_list_df, duplicated_grade_report_df, 
                                 how='inner', on=['usim'])
duplicated_grade_report_df = duplicated_grade_report_df.astype({ 'ts_weekday' : str })
duplicated_grade_report_df
```
